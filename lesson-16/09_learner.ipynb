````python
class Learner:
    def __init__(self, model, dls, loss_func, lr, opt_func=optim.SGD): fc.store_attr()

    def one_batch(self):
        self.xb,self.yb = to_device(self.batch)
        self.preds = self.model(self.xb)
        self.loss = self.loss_func(self.preds, self.yb)
        if self.model.training:
            self.loss.backward()
            self.opt.step()
            self.opt.zero_grad()
        with torch.no_grad(): self.calc_stats()

    def calc_stats(self):
        acc = (self.preds.argmax(dim=1)==self.yb).float().sum()
        self.accs.append(acc)
        n = len(self.xb)
        self.losses.append(self.loss*n)
        self.ns.append(n)

    def one_epoch(self, train):
        self.model.training = train
        dl = self.dls.train if train else self.dls.valid
        for self.num,self.batch in enumerate(dl): self.one_batch()
        n = sum(self.ns)
        print(self.epoch, self.model.training, sum(self.losses).item()/n, sum(self.accs).item()/n)
    
    def fit(self, n_epochs):
        self.accs,self.losses,self.ns = [],[],[]
        self.model.to(def_device)
        self.opt = self.opt_func(self.model.parameters(), self.lr)
        self.n_epochs = n_epochs
        for self.epoch in range(n_epochs):
            self.one_epoch(True)
            with torch.no_grad(): self.one_epoch(False)
````

## Basic Callbacks Learner

````python
#|export
class CancelFitException(Exception): pass
class CancelBatchException(Exception): pass
class CancelEpochException(Exception): pass
````

````python
#|export
class Callback(): order = 0
````

````python
#|export
def run_cbs(cbs, method_nm, learn=None):
    for cb in sorted(cbs, key=attrgetter('order')):
        method = getattr(cb, method_nm, None)
        if method is not None: method(learn)
````

````python
class CompletionCB(Callback):
    def before_fit(self, learn): self.count = 0
    def after_batch(self, learn): self.count += 1
    def after_fit(self, learn): print(f'Completed {self.count} batches')
````

````python
cbs = [CompletionCB()]
run_cbs(cbs, 'before_fit')
run_cbs(cbs, 'after_batch')
run_cbs(cbs, 'after_fit')
````

````python
class Learner():
    def __init__(self, model, dls, loss_func, lr, cbs, opt_func=optim.SGD): fc.store_attr()

    def one_batch(self):
        self.preds = self.model(self.batch[0])
        self.loss = self.loss_func(self.preds, self.batch[1])
        if self.model.training:
            self.loss.backward()
            self.opt.step()
            self.opt.zero_grad()

    def one_epoch(self, train):
        self.model.train(train)
        self.dl = self.dls.train if train else self.dls.valid
        try:
            self.callback('before_epoch')
            for self.iter,self.batch in enumerate(self.dl):
                try:
                    self.callback('before_batch')
                    self.one_batch()
                    self.callback('after_batch')
                except CancelBatchException: pass
            self.callback('after_epoch')
        except CancelEpochException: pass
    
    def fit(self, n_epochs):
        self.n_epochs = n_epochs
        self.epochs = range(n_epochs)
        self.opt = self.opt_func(self.model.parameters(), self.lr)
        try:
            self.callback('before_fit')
            for self.epoch in self.epochs:
                self.one_epoch(True)
                self.one_epoch(False)
            self.callback('after_fit')
        except CancelFitException: pass

    def callback(self, method_nm): run_cbs(self.cbs, method_nm, self)
````

## Metrics

````python
class Metric:
    def __init__(self): self.reset()
    def reset(self): self.vals,self.ns = [],[]
    def add(self, inp, targ=None, n=1):
        self.last = self.calc(inp, targ)
        self.vals.append(self.last)
        self.ns.append(n)
    @property
    def value(self):
        ns = tensor(self.ns)
        return (tensor(self.vals)*ns).sum()/ns.sum()
    def calc(self, inps, targs): return inps
````

````python
class Accuracy(Metric):
    def calc(self, inps, targs): return (inps==targs).float().mean()
````

````python
#|export
def to_cpu(x):
    if isinstance(x, Mapping): return {k:to_cpu(v) for k,v in x.items()}
    if isinstance(x, list): return [to_cpu(o) for o in x]
    if isinstance(x, tuple): return tuple(to_cpu(list(x)))
    res = x.detach().cpu()
    return res.float() if res.dtype==torch.float16 else res
````

````python
#|export
class MetricsCB(Callback):
    def __init__(self, *ms, **metrics):
        for o in ms: metrics[type(o).__name__] = o
        self.metrics = metrics
        self.all_metrics = copy(metrics)
        self.all_metrics['loss'] = self.loss = Mean()

    # 默认的日志输出方法，直接 print dict
    # 当 ProgressCB 存在时，会被替换为 ProgressCB._log 以渲染 HTML 表格
    def _log(self, d): print(d)
    def before_fit(self, learn): learn.metrics = self
    def before_epoch(self, learn): [o.reset() for o in self.all_metrics.values()]

    def after_epoch(self, learn):
        # 组装一行指标数据: {accuracy: '0.624', loss: '1.134', epoch: 0, train: 'train'}
        log = {k:f'{v.compute():.3f}' for k,v in self.all_metrics.items()}
        log['epoch'] = learn.epoch
        log['train'] = 'train' if learn.model.training else 'eval'
        # 调用 _log 输出（可能已被 ProgressCB 替换为表格渲染）
        self._log(log)

    def after_batch(self, learn):
        x,y,*_ = to_cpu(learn.batch)
        for m in self.metrics.values(): m.update(to_cpu(learn.preds), y)
        self.loss.update(to_cpu(learn.loss), weight=len(x))
````

````python
#|export
class DeviceCB(Callback):
    def __init__(self, device=def_device): fc.store_attr()
    def before_fit(self, learn):
        if hasattr(learn.model, 'to'): learn.model.to(self.device)
    def before_batch(self, learn): learn.batch = to_device(learn.batch, device=self.device)
````

## Flexible learner

````python
class Learner():
    def __init__(self, model, dls=(0,), loss_func=F.mse_loss, lr=0.1, cbs=None, opt_func=optim.SGD):
        cbs = fc.L(cbs)
        fc.store_attr()

    @contextmanager
    def cb_ctx(self, nm):
        try:
            self.callback(f'before_{nm}')
            yield
            self.callback(f'after_{nm}')
        except globals()[f'Cancel{nm.title()}Exception']: pass
        finally: self.callback(f'cleanup_{nm}')
                
    def one_epoch(self, train):
        self.model.train(train)
        self.dl = self.dls.train if train else self.dls.valid
        with self.cb_ctx('epoch'):
            for self.iter,self.batch in enumerate(self.dl):
                with self.cb_ctx('batch'):
                    self.predict()
                    self.get_loss()
                    if self.training:
                        self.backward()
                        self.step()
                        self.zero_grad()
    
    def fit(self, n_epochs=1, train=True, valid=True, cbs=None, lr=None):
        cbs = fc.L(cbs)
        # `add_cb` and `rm_cb` were added in lesson 18
        for cb in cbs: self.cbs.append(cb)
        try:
            self.n_epochs = n_epochs
            self.epochs = range(n_epochs)
            self.opt = self.opt_func(self.model.parameters(), self.lr if lr is None else lr)
            # with self.cb_ctx('fit'):
            #    ↓
            # 
            # __enter__()
            #    ↓
            # before_fit
            #    ↓
            # 执行 with 代码块
            #    ↓
            # __exit__()
            #    ↓
            # after_fit (如果没异常)
            #    ↓
            # cleanup_fit (一定执行)
            with self.cb_ctx('fit'):
                for self.epoch in self.epochs:
                    if train: self.one_epoch(True)
                    if valid: torch.no_grad()(self.one_epoch)(False)
        finally:
            for cb in cbs: self.cbs.remove(cb)

    def __getattr__(self, name):
        if name in ('predict','get_loss','backward','step','zero_grad'): return partial(self.callback, name)
        raise AttributeError(name)

    def callback(self, method_nm): run_cbs(self.cbs, method_nm, self)
    
    @property
    def training(self): return self.model.training
````

````python
#|export
class TrainCB(Callback):
    def __init__(self, n_inp=1): self.n_inp = n_inp
    def predict(self, learn): learn.preds = learn.model(*learn.batch[:self.n_inp])
    def get_loss(self, learn): learn.loss = learn.loss_func(learn.preds, *learn.batch[self.n_inp:])
    def backward(self, learn): learn.loss.backward()
    def step(self, learn): learn.opt.step()
    def zero_grad(self, learn): learn.opt.zero_grad()
````

NB: I added `self.n_inp` after the lesson. This allows us to train models with more than one input or output.

````python
#|export
class ProgressCB(Callback):
    order = MetricsCB.order+1  # 确保在 MetricsCB 之后执行，这样 before_fit 时 learn.metrics 已存在
    def __init__(self, plot=False): self.plot = plot
    def before_fit(self, learn):
        # master_bar 包装 epochs range，提供总体进度条 + 表格 + 图表的容器
        learn.epochs = self.mbar = master_bar(learn.epochs)
        self.first = True
        # 猴子补丁: 把 MetricsCB._log 替换为自己的 _log，接管指标输出
        # 这样 MetricsCB.after_epoch 调用 self._log 时，实际执行的是下面的 _log
        if hasattr(learn, 'metrics'): learn.metrics._log = self._log
        self.losses = []
        self.val_losses = []

    def _log(self, d):
        # 由 MetricsCB.after_epoch 调用，d 是 {accuracy, loss, epoch, train} 的 dict
        if self.first:
            # 第一次调用: 写表头，如 ['accuracy', 'loss', 'epoch', 'train']
            self.mbar.write(list(d), table=True)
            self.first = False
        # 每次调用: 写数据行，如 ['0.624', '1.134', 0, 'train']
        self.mbar.write(list(d.values()), table=True)

    def before_epoch(self, learn):
        # 用 progress_bar 包装 DataLoader，迭代 batch 时显示进度条
        # leave=False: epoch 结束后进度条消失; parent=self.mbar: 嵌套在 master_bar 下
        learn.dl = progress_bar(learn.dl, leave=False, parent=self.mbar)

    def after_batch(self, learn):
        # 在进度条右侧显示当前 batch 的 loss 值
        learn.dl.comment = f'{learn.loss:.3f}'
        if self.plot and hasattr(learn, 'metrics') and learn.training:
            self.losses.append(learn.loss.item())
            if self.val_losses:
                # update_graph 接收两组数据: [[train_x, train_y], [val_x, val_y]]
                # train_x: [0,1,2,...] 每个 batch 的序号
                # train_y: 每个 batch 的训练 loss
                # val_x: 每个 epoch 结束对应的 batch 位置，如 [59, 118]
                # val_y: 每个 epoch 的验证 loss
                graph_data = [[fc.L.range(self.losses), self.losses],[fc.L.range(learn.epoch).map(lambda x: (x+1)*len(learn.dls.train)), self.val_losses]]
                # 调试: 每个 epoch 的第一个 batch 时打印参数
                if learn.iter == 0:
                    print(f"\n--- after_batch update_graph 参数 (epoch={learn.epoch}, batch={learn.iter}) ---")
                    print(f"  训练 loss 曲线: x={list(graph_data[0][0][:5])}...{list(graph_data[0][0][-3:])} (共{len(graph_data[0][0])}点)")
                    print(f"                  y={graph_data[0][1][:5]}...{graph_data[0][1][-3:]} (共{len(graph_data[0][1])}点)")
                    print(f"  验证 loss 散点: x={list(graph_data[1][0])}")
                    print(f"                  y={graph_data[1][1]}")
                self.mbar.update_graph(graph_data)
    
    def after_epoch(self, learn): 
        if not learn.training:
            if self.plot and hasattr(learn, 'metrics'): 
                # 验证 epoch 结束，记录验证 loss 并更新图表
                self.val_losses.append(learn.metrics.all_metrics['loss'].compute())
                self.mbar.update_graph([[fc.L.range(self.losses), self.losses],[fc.L.range(learn.epoch+1).map(lambda x: (x+1)*len(learn.dls.train)), self.val_losses]])
````

## Updated versions since the lesson

After the lesson we noticed that `contextlib.context_manager` has a surprising "feature" which doesn't let us raise an exception before the `yield`. Therefore we've replaced the context manager with a decorator in this updated version of `Learner`. We have also added a few more callbacks in `one_epoch()`.

````python
#|export
class with_cbs:
    def __init__(self, nm): self.nm = nm
    def __call__(self, f):
        def _f(o, *args, **kwargs):
            try:
                o.callback(f'before_{self.nm}')
                f(o, *args, **kwargs)
                o.callback(f'after_{self.nm}')
            except globals()[f'Cancel{self.nm.title()}Exception']: pass
            finally: o.callback(f'cleanup_{self.nm}')
        return _f
````

````python
#|export
class Learner():
    def __init__(self, model, dls=(0,), loss_func=F.mse_loss, lr=0.1, cbs=None, opt_func=optim.SGD):
        # 将 cbs 转为 fastcore 的 L 列表（支持 None 输入，自动变为空列表）
        cbs = fc.L(cbs)
        # store_attr() 自动将所有参数存为 self.model, self.dls, self.loss_func 等实例属性
        fc.store_attr()

    @with_cbs('batch')
    def _one_batch(self):
        # with_cbs('batch') 装饰器会在执行前后自动调用 before_batch / after_batch 回调
        # 并捕获 CancelBatchException 以支持提前终止当前 batch
        self.predict()              # 触发 'predict' 回调 → 由 TrainCB.predict 执行前向传播
        self.callback('after_predict')  # 额外的钩子点，可用于在预测后做处理（如 mixup）
        self.get_loss()             # 触发 'get_loss' 回调 → 由 TrainCB.get_loss 计算损失
        self.callback('after_loss')     # 额外的钩子点，可用于在损失计算后做处理
        if self.training:           # 只在训练模式下执行反向传播和参数更新
            self.backward()         # 触发 'backward' 回调 → 由 TrainCB.backward 执行 loss.backward()
            self.callback('after_backward')  # 钩子点：可用于梯度裁剪等
            self.step()             # 触发 'step' 回调 → 由 TrainCB.step 执行 opt.step()
            self.callback('after_step')      # 钩子点：可用于学习率调度等
            self.zero_grad()        # 触发 'zero_grad' 回调 → 由 TrainCB.zero_grad 清零梯度

    @with_cbs('epoch')
    def _one_epoch(self):
        # with_cbs('epoch') 装饰器会在执行前后自动调用 before_epoch / after_epoch 回调
        # 并捕获 CancelEpochException 以支持提前终止当前 epoch
        for self.iter,self.batch in enumerate(self.dl): self._one_batch()

    def one_epoch(self, training):
        # 设置模型为训练/评估模式（影响 dropout、batchnorm 等层的行为）
        self.model.train(training)
        # 根据训练/验证选择对应的 DataLoader
        self.dl = self.dls.train if training else self.dls.valid
        self._one_epoch()

    @with_cbs('fit')
    def _fit(self, train, valid):
        # with_cbs('fit') 装饰器会在执行前后自动调用 before_fit / after_fit 回调
        # 并捕获 CancelFitException 以支持提前终止整个训练
        for self.epoch in self.epochs:
            if train: self.one_epoch(True)
            # 验证时用 torch.no_grad() 包装，禁用梯度计算以节省内存和加速
            if valid: torch.no_grad()(self.one_epoch)(False)

    def fit(self, n_epochs=1, train=True, valid=True, cbs=None, lr=None):
        cbs = fc.L(cbs)
        # 临时添加额外的回调（如 LRFinderCB），训练结束后会在 finally 中移除
        for cb in cbs: self.cbs.append(cb)
        try:
            self.n_epochs = n_epochs
            self.epochs = range(n_epochs)
            if lr is None: lr = self.lr
            # 创建优化器，如果 opt_func 为 None 则跳过（某些特殊场景）
            if self.opt_func: self.opt = self.opt_func(self.model.parameters(), lr)
            self._fit(train, valid)
        finally:
            # 确保临时回调被移除，即使训练中途异常退出
            for cb in cbs: self.cbs.remove(cb)

    def __getattr__(self, name):
        # 魔法方法：当访问 predict/get_loss/backward/step/zero_grad 时
        # 自动转化为 self.callback(name)，即触发对应名称的回调
        # 这样 TrainCB 等回调只需定义同名方法即可被调用
        if name in ('predict','get_loss','backward','step','zero_grad'): return partial(self.callback, name)
        raise AttributeError(name)

    def callback(self, method_nm): run_cbs(self.cbs, method_nm, self)
    
    @property
    def training(self): return self.model.training
````

## TrainLearner and MomentumLearner

````python
#|export
class TrainLearner(Learner):
    def predict(self): self.preds = self.model(self.batch[0])
    def get_loss(self): self.loss = self.loss_func(self.preds, self.batch[1])
    def backward(self): self.loss.backward()
    def step(self): self.opt.step()
    def zero_grad(self): self.opt.zero_grad()
````

````python
#|export
class MomentumLearner(TrainLearner):
    def __init__(self, model, dls, loss_func, lr=None, cbs=None, opt_func=optim.SGD, mom=0.85):
        self.mom = mom
        super().__init__(model, dls, loss_func, lr, cbs, opt_func)

    def zero_grad(self):
        with torch.no_grad():
            for p in self.model.parameters(): p.grad *= self.mom
````

## LRFinderCB

````python
class LRFinderCB(Callback):
    def __init__(self, lr_mult=1.3): fc.store_attr()
    
    def before_fit(self, learn):
        self.lrs,self.losses = [],[]
        self.min = math.inf

    def after_batch(self, learn):
        if not learn.training: raise CancelEpochException()
        self.lrs.append(learn.opt.param_groups[0]['lr'])
        loss = to_cpu(learn.loss)
        self.losses.append(loss)
        if loss < self.min: self.min = loss
        if loss > self.min*3: raise CancelFitException()
        for g in learn.opt.param_groups: g['lr'] *= self.lr_mult
````

[ExponentialLR](https://pytorch.org/docs/stable/generated/torch.optim.lr_scheduler.ExponentialLR.html#torch.optim.lr_scheduler.ExponentialLR)

````python
#|export
class LRFinderCB(Callback):
    def __init__(self, gamma=1.3, max_mult=3): fc.store_attr()
    
    def before_fit(self, learn):
        self.sched = ExponentialLR(learn.opt, self.gamma)
        self.lrs,self.losses = [],[]
        self.min = math.inf

    def after_batch(self, learn):
        if not learn.training: raise CancelEpochException()
        self.lrs.append(learn.opt.param_groups[0]['lr'])
        loss = to_cpu(learn.loss)
        self.losses.append(loss)
        if loss < self.min: self.min = loss
        if math.isnan(loss) or (loss > self.min*self.max_mult):
            raise CancelFitException()
        self.sched.step()

    def cleanup_fit(self, learn):
        plt.plot(self.lrs, self.losses)
        plt.xscale('log')
````

````python
cbs = [DeviceCB()]
learn = MomentumLearner(get_model(), dls, F.cross_entropy, lr=1e-5, cbs=cbs)
learn.fit(3, cbs=LRFinderCB())
````

````python
#|export
@fc.patch
def lr_find(self:Learner, gamma=1.3, max_mult=3, start_lr=1e-5, max_epochs=10):
    self.fit(max_epochs, lr=start_lr, cbs=LRFinderCB(gamma=gamma, max_mult=max_mult))
````
